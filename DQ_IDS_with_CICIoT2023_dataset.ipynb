## **RL-Based Adaptive IDS Implementation (DQN)**

```python
# Import Necessary Libraries
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import random
import matplotlib.pyplot as plt
from collections import deque
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# ======================================
# 🚀 Step 1: Data Preprocessing (Ensuring Consistency)
# ======================================

# Assume 'df' is the preprocessed dataset from earlier steps
df = df.sample(frac=1).reset_index(drop=True)  # Shuffle dataset

# Encode Labels (Benign=0, Attack=1)
df['label'] = df['label'].apply(lambda x: 0 if x == 'BenignTraffic' else 1)

# Feature Selection
features = df.drop(columns=['label'])
labels = df['label']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Convert data to PyTorch tensors
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.long)
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.long)

# ======================================
# 🚀 Step 2: Define RL-IDS Environment
# ======================================

class NetworkIDS:
    def __init__(self, input_dim, action_dim):
        self.state_size = input_dim  # Feature count
        self.action_size = action_dim  # Attack (1) or Benign (0)
        self.memory = deque(maxlen=10000)  # Experience Replay Memory
        self.gamma = 0.95  # Discount factor
        self.epsilon = 1.0  # Initial exploration rate
        self.epsilon_min = 0.01  # Minimum exploration rate
        self.epsilon_decay = 0.995  # Decay rate for exploration
        self.learning_rate = 0.001

        # Neural Network Model (Deep Q-Network)
        self.model = nn.Sequential(
            nn.Linear(self.state_size, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, self.action_size),
        )

        self.optimizer = optim.Adam(self.model.parameters(), lr=self.learning_rate)
        self.criterion = nn.MSELoss()

    # Store Experience (Replay Buffer)
    def store_experience(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    # Select Action (Exploration vs Exploitation)
    def select_action(self, state):
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)  # Random action (Exploration)
        state_tensor = torch.tensor(state, dtype=torch.float32)
        with torch.no_grad():
            q_values = self.model(state_tensor)  # Get Q-values
        return torch.argmax(q_values).item()  # Exploitation (Best Action)

    # Train RL Model
    def train_model(self, batch_size=64):
        if len(self.memory) < batch_size:
            return None, None  # Ensure safe return values

        # Sample a batch of experiences
        batch = random.sample(self.memory, batch_size)
        losses = []
        rewards = []

        for state, action, reward, next_state, done in batch:
            state_tensor = torch.tensor(state, dtype=torch.float32)
            next_state_tensor = torch.tensor(next_state, dtype=torch.float32)
            reward_tensor = torch.tensor(reward, dtype=torch.float32)

            target = reward_tensor
            if not done:
                with torch.no_grad():
                    target += self.gamma * torch.max(self.model(next_state_tensor))

            output = self.model(state_tensor)[action]  # Q-value for action taken
            loss = self.criterion(output, target)

            self.optimizer.zero_grad()
            loss.backward()
            self.optimizer.step()

            losses.append(loss.item())
            rewards.append(reward)

        # Decrease exploration rate
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

        return np.mean(losses), np.mean(rewards)  # Return average loss and reward

# ======================================
# 🚀 Step 3: Train RL-IDS with Performance Tracking
# ======================================

input_dim = X_train.shape[1]  # Number of features
action_dim = 2  # Benign (0) or Attack (1)

agent = NetworkIDS(input_dim, action_dim)

num_episodes = 1500  # Training iterations
batch_size = 64

loss_history = []
reward_history = []

for episode in range(num_episodes):
    state = X_train.iloc[random.randint(0, len(X_train)-1)].values
    done = False
    episode_rewards = []

    while not done:
        action = agent.select_action(state)
        next_state = X_train.iloc[random.randint(0, len(X_train)-1)].values
        reward = 1 if action == y_train.iloc[random.randint(0, len(y_train)-1)] else -1  # Reward for correct classification
        episode_rewards.append(reward)
        done = True  # Each sample is a one-step episode

        agent.store_experience(state, action, reward, next_state, done)
        avg_loss, avg_reward = agent.train_model(batch_size)

    if avg_loss is not None and avg_reward is not None:  # Avoid NoneType formatting issue
        loss_history.append(avg_loss)
        reward_history.append(np.mean(episode_rewards))

        if episode % 100 == 0:
            print(f"Episode {episode}, Epsilon: {agent.epsilon:.3f}, Avg Loss: {avg_loss:.4f}, Avg Reward: {np.mean(episode_rewards):.4f}")
    else:
        loss_history.append(0)  # Default value before training starts
        reward_history.append(0)

print("Training Complete!")

# ======================================
# 🚀 Step 4: Evaluate RL-IDS
# ======================================

correct = 0
total = len(X_test)

for i in range(total):
    state = X_test.iloc[i].values
    action = agent.select_action(state)
    if action == y_test.iloc[i]:
        correct += 1

accuracy = correct / total
print(f"RL-IDS Accuracy: {accuracy * 100:.2f}%")
```

```python
# ======================================
# 🚀 Step 5: Plot Training Performance Metrics
# ======================================

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 3))

# Plot Loss Trend
plt.subplot(1, 1, 1)
plt.plot(loss_history, label="Training Loss", color="red")
plt.xlabel("Episodes")
plt.ylabel("Loss")
plt.title("Loss Trend Over Training")
plt.legend()
plt.grid(True)  # Enable grid

plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# ======================================
# 🚀 Step 4: Evaluate RL-IDS with Metrics
# ======================================

y_pred = []  # Predicted labels
y_true = y_test.tolist()  # Actual labels

# Predict labels for test data
for i in range(len(X_test)):
    state = X_test.iloc[i].values
    action = agent.select_action(state)
    y_pred.append(action)

# Convert predictions and true labels to numpy arrays
y_pred = np.array(y_pred)
y_true = np.array(y_true)

# Compute Evaluation Metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
conf_matrix = confusion_matrix(y_true, y_pred)

# Compute False Positive Rate (FPR) and False Negative Rate (FNR)
tn, fp, fn, tp = conf_matrix.ravel()
fpr = fp / (fp + tn)  # False Positive Rate
fnr = fn / (fn + tp)  # False Negative Rate

# Print Evaluation Metrics
print("\n========= RL-IDS Evaluation Metrics =========")
print(f"✅ Accuracy: {accuracy * 100:.2f}%")
print(f"✅ Precision: {precision:.4f}")
print(f"✅ Recall (Sensitivity): {recall:.4f}")
print(f"✅ F1 Score: {f1:.4f}")
print(f"✅ False Positive Rate (FPR): {fpr:.4f}")
print(f"✅ False Negative Rate (FNR): {fnr:.4f}")

# Print Confusion Matrix
print("\n📊 Confusion Matrix:")
print(conf_matrix)
```
